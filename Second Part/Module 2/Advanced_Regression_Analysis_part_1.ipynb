# Polinomial Models

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Sample dataset
data = {
    'Study_Hours': [1, 2, 3, 4, 5, 6],
    'Exam_Score': [50, 65, 75, 80, 85, 88]
}

# Create a DataFrame from the dataset
df = pd.DataFrame(data)
```

```python
# Extract the features (Study_Hours) and target (Exam_Score)
feature = np.array(data['Study_Hours']).reshape(-1, 1)
target = np.array(data['Exam_Score'])
```

```python
# Fit linear regression for comparison
lin_model = LinearRegression()
lin_model.fit(feature, target)
```

```python
# Fit polynomial regression
pol_feature = PolynomialFeatures()
pol_feature = pol_feature.fit_transform(feature)
pol_model = LinearRegression()
pol_model.fit(pol_feature, target)
```

```python
lin_y = lin_model.predict(feature)
pol_y = pol_model.predict(pol_feature)

# Plot results
plt.scatter(feature, target)
plt.plot(feature, lin_y, '--', color='red', label='Linear')
plt.plot(feature, pol_y, '--', color='green', label='Polynomial')
plt.legend()
plt.show()

# Evaluate the models
print("\nModels metrics:")
print(f"Linear regression MSE: {mean_squared_error(target, lin_y)}")
print(f"Polynomial regression MSE: {mean_squared_error(target, pol_y)}")
```

# Exponential regression

```python
# Sample dataset
data = {
    'Date': pd.date_range(start='2022-01-01', periods=10),
    'Engaged_Users': [100, 150, 225, 338, 507, 761, 1142, 1713, 2570, 3855]
}

# Create a DataFrame from the dataset
data = pd.DataFrame(data)

# Extract the features (Date) and target (Engaged_Users)
feature = np.array(data['Date'].astype(int)).reshape(-1, 1)
target = data['Engaged_Users']

# Visualize the data
plt.scatter(feature, target)

# Fit linear regression for comparison
lin_model = LinearRegression()
lin_model.fit(feature, target)
lin_y = lin_model.predict(feature)
plt.plot(feature, lin_y, '--', color='red', label='Linear')

# Fit exponential regression
exp_target = np.log(target)
exp_model = LinearRegression()
exp_model.fit(feature, exp_target)
exp_y = exp_model.predict(feature)
exp_y = np.exp(exp_y)
plt.plot(feature, exp_y, '--', color='blue', label='Exponential')

# Display plot
plt.legend()
plt.show()

# Evaluate the models
print("\nModels metrics:")
print(f"Linear regression MSE: {mean_squared_error(target, lin_y)}")
print(f"Exponential regression MSE: {mean_squared_error(target, exp_y)}")
```

# Power models

```python
# Import libraries

# Create a DataFrame with the sample data
data = {
    'User_ID': range(1, 11),
    'Followers': np.arange(10, 110, 10),
    'Likes': [25, 50, 100, 200, 350, 550, 800, 1100, 1450, 1850],
    'Shares': [5, 10, 20, 40, 70, 110, 160, 220, 290, 370],
    'Comments': [2, 4, 8, 16, 28, 44, 64, 88, 116, 148]
}
df = pd.DataFrame(data)
df.drop(columns=['User_ID'])

# Get feature and target
feature = np.array(df['Followers']).reshape(-1, 1)
target = df['Likes']

# Visualize data
plt.scatter(feature, target)

# Perform power regression for Likes as an example
log_feature = np.log(feature)
log_target = np.log(target)
lin_model = LinearRegression()
lin_model.fit(log_feature, log_target)
lin_y = np.exp(lin_model.predict(log_feature))
plt.plot(feature, lin_y, '--', color='pink', label='Fitted curve')

# Generate a range of Followers values for the fitted curve
new_x = np.arange(55, 75, 5).reshape(-1, 1)

# Calculate the predicted Likes using the fitted parameters
lin_y = np.exp(lin_model.predict(np.log(new_x)))
plt.plot(new_x, lin_y, '--', color='red', label='Prediction')

# Create a plot to visualize the relationship
plt.legend()
plt.show()
```
